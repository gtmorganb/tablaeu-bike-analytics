```python
import os 
import csv 
import pandas as pd
```

```python
file_to_load = "Resources/july.csv"
file_to_load2 = "Resources/aug.csv"
file_to_load3 = "Resources/sept.csv"
file_to_load4 = "Resources/octo.csv"
file_to_load5 = "Resources/nov.csv"
file_to_load6 = "Resources/dec.csv"


july = pd.read_csv(file_to_load)
aug = pd.read_csv(file_to_load2)
sept = pd.read_csv(file_to_load3)
octo = pd.read_csv(file_to_load4)
nov = pd.read_csv(file_to_load5)
dec = pd.read_csv(file_to_load6)
```

```python
jul_dec = [july, aug, sept, octo, nov, dec]
```

```python
yr7_12 = pd.concat(jul_dec)
```

```python
yr7_12.head()
```

```python
yr7_12.to_csv("2020 July to Dec.csv")
```

```python
year_totals = {"2020 July-Dec": ['July','August','September','October','November','December'], 
'Number of Trips': [len(july),len(aug),len(sept),len(octo),len(nov),len(dec)]}  
```

```python
year_totals_df = pd.DataFrame(year_totals)
year_totals_df.head()
```

```python
year_totals_df.to_csv("Trips per Month 2020 July to Dec.csv")
```
